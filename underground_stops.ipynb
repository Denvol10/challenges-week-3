```python
import pandas as pd
import json
from geopy import distance
```

```python
stops = pd.read_csv(open('bus_stops.csv', 'r', encoding='cp1251'), sep=';') # четния базы данных с остановками транспорта
```

```python
underground = pd.read_json(open('data-397-2019-05-16.json', 'r', encoding='cp1251')) # чтение базы со станциями метро
```

```python
# измеряем расстояние между двумя точками, используя модуль geopy
def distances(x1, y1, x2, y2):
    point_1 = (float(x1), float(y1))
    point_2 = (float(x2), float(y2))
    return distance.distance(point_1, point_2).km
```

```python
# берем координаты станции метро и измеряем расстояние до остановок (модуль geopy на большом количестве данных работает медленно
# возможно следовало определять расстояния приблеженно)
dict_underground = {}
for under_index, under_row in underground.iterrows():
    x1 = under_row['Longitude_WGS84']
    y1 = under_row['Latitude_WGS84']
    name_underground = under_row['NameOfStation']
    list_stops = []
    for stops_index, stops_row in stops.iterrows():
        x2 = stops_row['Longitude_WGS84']
        y2 = stops_row['Latitude_WGS84']
        name_stops = stops_row['StationName']
        if x2 == 'Longitude_WGS84' or y2 == 'Latitude_WGS84':
            pass
        else:
            if distances(x1, y1, x2, y2) <= 0.5:
                if not name_stops in list_stops: # считаем разными остановками только те, которые имеют разные имена
                    list_stops.append(name_stops)
    dict_underground[name_underground] = list_stops
```

```python
dict_underground
```

```python
max_count_stops = 0
for station, stops in dict_underground.items():
    if max_count_stops < len(stops):
        max_count_stops = len(stops)
        max_station = station
print('Максимальное количество остановок в радиусе 0,5 км имеет станция {}: {} шт'.format(max_station, max_count_stops))
```
