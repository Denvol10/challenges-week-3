```python
import pandas as pd
import json
from geopy import distance
```

```python
stops = pd.read_csv(open('bus_stops.csv', 'r', encoding='cp1251'), sep=';') # четния базы данных с остановками транспорта
```

```python
underground = pd.read_json(open('data-397-2019-05-16.json', 'r', encoding='cp1251')) # чтение базы со станциями метро
```

```python
underground.drop_duplicates(subset='NameOfStation', inplace=True) # удаление строк, относящихся к одной и той же станции
# поскольку одна станция метро имеет несколько входов(выходов)
```

```python
# измеряем расстояние между двумя точками, используя модуль geopy
def distances(x1, y1, x2, y2):
    point_1 = (float(x1), float(y1))
    point_2 = (float(x2), float(y2))
    return distance.distance(point_1, point_2).km
```

```python
# берем координаты станции метро и измеряем расстояние до остановок (модуль geopy на большом количестве данных работает медленно
# возможно следовало определять расстояния приблеженно)
count_max = 0
for under_index, under_coordinats in underground.iterrows():
    x1 = under_coordinats['Longitude_WGS84']
    y1 = under_coordinats['Latitude_WGS84']
    count_stops = 0
    for stops_index, stops_coordinats in stops.iterrows():
        x2 = stops_coordinats['Longitude_WGS84']
        y2 = stops_coordinats['Latitude_WGS84']
        if x2 == 'Longitude_WGS84' or y2 == 'Latitude_WGS84':
            pass
        else:
            if distances(x1, y1, x2, y2) <= 0.5:
                count_stops += 1
    if count_max < count_stops:
        count_max = count_stops
        under_max_index = under_index      
                
```

```python
print('Больше всего остановок у станции метро в радиусе 0,5 км имеет станция {}: {} остановки'.format(underground.loc[under_max_index]['NameOfStation'], count_max))
```
