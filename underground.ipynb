```python
import pandas as pd
import json
from datetime import datetime, date, timedelta
```

```python
# чтение базы данных в формате json
underground = pd.read_json(open('data-397-2019-05-16.json', 'r', encoding='cp1251'))
```

```python
# фильтр для исключения станций на которых не производится ремонт
fn_filter = underground['RepairOfEscalators'].map(lambda x: len(x) != 0)
```

```python
df_rerair_of_Escal = underground[fn_filter]
```

```python
# функция, которая проверяет, попадает ли число сегодня в интервал, включая границы интервала
def period_of_time(date_string):
    date_list = date_string.split('-')
    date_start = datetime.strptime(date_list[0], '%d.%m.%Y')
    date_finish = datetime.strptime(date_list[1], '%d.%m.%Y')
    date_now = datetime.now() # дата сегодня
    #date_now = datetime.strptime('07.02.2019', '%d.%m.%Y') дата для проверки функции, поскольку date_now может
    # никогда не попадать в интервал из-за устаревшей базы данных
    delta_1 = (date_start - date_now).days
    delta_2 = (date_finish - date_now).days
    if delta_1 < 0 and delta_2 > 0:
        return True
    elif delta_1 == -1 or delta_2 == -1:
        return True
    else:
        return False
```

```python
# включаем в словарь станции на которых производится сегодня ремонт
dict_stations_of_Repair = {}
for index, row in df_rerair_of_Escal.iterrows():
    dates_of_Repair = row['RepairOfEscalators'][0]['RepairOfEscalators']
    if period_of_time(dates_of_Repair):
        dict_stations_of_Repair[row['Name']] = dates_of_Repair
```

```python
if dict_stations_of_Repair == {}:
    print('Эскалатор работает на всех станциях')
else:
    print('Сегодня прозводится ремонт на станциях:')
    for station in dict_stations_of_Repair.keys():
        print(station)
```
